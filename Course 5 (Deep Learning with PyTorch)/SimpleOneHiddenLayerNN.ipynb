<h2 id="Model">Neural Network Module and Training Function</h2> 


Define the activations and the output of the first linear layer as an attribute. Note that this is not good practice. 


```python
import torch.nn as nn

class Net(nn.Module):
    """
    One-hidden-layer network:
      x -> Linear(D_in -> H) -> sigmoid -> Linear(H -> D_out) -> sigmoid -> yhat

    This version intentionally stores intermediate tensors (l1, a1, l2)
    on `self` for teaching/visualization purposes.
    """

    def __init__(self, D_in, H, D_out):
        super(Net, self).__init__()

        # These are the *actual trainable layers* (standard PyTorch practice).
        # - linear1 maps input features -> hidden units
        # - linear2 maps hidden units -> output units
        self.linear1 = nn.Linear(D_in, H)
        self.linear2 = nn.Linear(H, D_out)

        # --- Teaching scaffolding (NOT good practice in production) ---
        # The authors store intermediate results as attributes so later cells
        # can inspect/plot what's happening inside the network (e.g., visualize
        # hidden activations, discuss pre-activation vs activation, etc.).
        #
        # Why this is normally avoided:
        # - Forward passes should be stateless (besides parameters).
        # - Storing per-batch tensors on `self` can cause confusing behavior when:
        #   * you run multiple forwards before reading them,
        #   * you use dataloaders with multiple workers,
        #   * you do inference concurrently,
        #   * you keep references that accidentally prolong GPU/CPU memory use.
        self.l1 = None  # hidden layer pre-activation (Linear1 output)
        self.a1 = None  # hidden layer activation (after sigmoid)
        self.l2 = None  # output layer pre-activation (Linear2 output)

    def forward(self, x):
        # Hidden linear transform (pre-activation)
        self.l1 = self.linear1(x)

        # Hidden activation
        # The notebook uses sigmoid here because it's easy to explain/plot.
        # In modern practice you'd more commonly see ReLU/GELU/etc.
        self.a1 = sigmoid(self.l1)

        # Output linear transform (pre-activation)
        # They store this as `self.l2` for inspection in later cells.
        self.l2 = self.linear2(self.a1)

        # --- Slightly odd / redundant part in the notebook ---
        # They recompute `linear2(self.a1)` again instead of reusing `self.l2`.
        # This isn't *wrong* (it's the same computation) but it's wasteful and
        # can confuse readers because it looks like `self.l2` matters but then
        # isn't actually used for yhat.
        #
        # A cleaner equivalent would be:
        #   yhat = sigmoid(self.l2)
        yhat = sigmoid(self.linear2(self.a1))

        return yhat
```

```python
#. Normal/clean production-style version (same math, no teaching scaffolding, no redundant compute):
class NetClean(nn.Module):
    def __init__(self, D_in, H, D_out):
        super().__init__()
        self.linear1 = nn.Linear(D_in, H)
        self.linear2 = nn.Linear(H, D_out)

    def forward(self, x):
        a1 = sigmoid(self.linear1(x))
        l2 = self.linear2(a1)
        yhat = sigmoid(l2)
        return yhat
```

Define the training function:


```python
# Define the training function

def train(Y, X, model, optimizer, criterion, epochs=1000):
    cost = []
    total=0
    for epoch in range(epochs):
        total=0
        for y, x in zip(Y, X):
            yhat = model(x)
            loss = criterion(yhat, y)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            #cumulative loss 
            total+=loss.item() 
        cost.append(total)
        if epoch % 300 == 0:    
            PlotStuff(X, Y, model, epoch, leg=True)
            plt.show()
            model(X)
            plt.scatter(model.a1.detach().numpy()[:, 0], model.a1.detach().numpy()[:, 1], c=Y.numpy().reshape(-1))
            plt.title('activations')
            plt.show()
    return cost
```

<h2 id="Makeup_Data">Make Some Data</h2>


```python
# Make some data

X = torch.arange(-20, 20, 1).view(-1, 1).type(torch.FloatTensor)
Y = torch.zeros(X.shape[0])
Y[(X[:, 0] > -4) & (X[:, 0] < 4)] = 1.0
```

<!--Empty Space for separating topics-->


<h2 id="Train">Define the Neural Network, Criterion Function, Optimizer and Train the Model</h2>


Create the Cross-Entropy loss function: 


```python
# The loss function

def criterion_cross(outputs, labels):
    out = -1 * torch.mean(labels * torch.log(outputs) + (1 - labels) * torch.log(1 - outputs))
    return out
```

Define the Neural Network, Optimizer, and Train the Model:


```python
# Train the model
# size of input 
D_in = 1
# size of hidden layer 
H = 2
# number of outputs 
D_out = 1
# learning rate 
learning_rate = 0.1
# create the model 
model = Net(D_in, H, D_out)
#optimizer 
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
#train the model usein
cost_cross = train(Y, X, model, optimizer, criterion_cross, epochs=1000)
#plot the loss
plt.plot(cost_cross)
plt.xlabel('epoch')
plt.title('cross entropy loss')
```

By examining the output of the  activation, you see by the 600th epoch that the data has been mapped to a linearly separable space.
